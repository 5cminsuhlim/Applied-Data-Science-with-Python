```python
import pandas as pd
import numpy as np

df = pd.read_csv('resources/week-3/datasets/cwurData.csv')
df.head()
```

```python
df_first = df[df['world_rank'] < 101]
df_second = df[df['world_rank'] < 201]
df_third = df[df['world_rank'] < 301]
df_other = df[df['world_rank'] > 300]

cumul = [df_first, df_second, df_third, df_other]

pd.concat(cumul, keys = ['First Tier Top University', 'Second Tier Top University', 'Third Tier Top University',
                         'Other Top University'])
```

```python
def create_category(ranking):
    if (ranking >= 1) & (ranking <= 100):
        return "First Top Tier University"
    elif (ranking >= 101) & (ranking <= 200):
        return "Secpnd Top Tier University"    
    elif (ranking >= 201) & (ranking <= 300):
        return "Third Top Tier University"
    return "Other Top University"

df['Rank_Level'] = df['world_rank'].apply(lambda df: create_category(df))

df.head()
```

```python
df.pivot_table(values = 'score', index = 'country', columns = 'Rank_Level', aggfunc = [np.mean]).head()
```

```python
df.pivot_table(values = 'score', index = 'country', columns = 'Rank_Level', aggfunc = [np.mean, np.max]).head()
```

```python
df.pivot_table(values = 'score', index = 'country', columns = 'Rank_Level', aggfunc = [np.mean, np.max],
               margins = True).head()
```

```python
new_df = df.pivot_table(values = 'score', index = 'country', columns = 'Rank_Level', aggfunc = [np.mean, np.max],
               margins = True)

print(new_df.index)
print(new_df.columns)
```
